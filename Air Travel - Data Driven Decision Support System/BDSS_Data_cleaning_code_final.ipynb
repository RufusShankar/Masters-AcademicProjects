```python
# Data Cleaning
def clean_and_combine_flight_data(file_paths):
    combined_df = pd.DataFrame()
    for file_path in file_paths:
        # Read the compressed CSV file
        df = pd.read_csv(file_path, compression='gzip')
#         print(df.shape)
        # Drop specific columns
        df.drop(['number','altitude_1','altitude_2'], axis=1, inplace=True)
        
        # Extract aircraft typecode
        df['typecode'] = df['typecode'].str.slice(0,1)
        
        # Calculate journey time in hours
        df['journey_time'] = (pd.to_datetime(df['lastseen']) - pd.to_datetime(df['firstseen'])).dt.total_seconds() / 3600.0
        df['journey_time'] = df['journey_time'].round(1)
        
        # Drop rows based on various conditions
        df.dropna(subset=['origin', 'destination'],inplace=True)
        df = df[df['origin'] != df['destination']]
        df.dropna(subset=['latitude_2','longitude_2'])
        df = df.drop(df[(df['callsign'].isna()) & (df['typecode'].isna())].index)

        #Append cleaned data to the combined DataFrame
        combined_df = pd.concat([combined_df, df], ignore_index=True)
    
    return combined_df
```
